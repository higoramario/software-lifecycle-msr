```python
from pydriller import Repository
import pandas as pd
import datetime as dt
```

```python
planilha_repos = pd.read_csv("CSV/links.csv")
```

```python
repos_list = planilha_repos['Links'].tolist()
```

```python
response = []
for commit in Repository(repos_list).traverse_commits():
    print(commit.project_name)
    response.append(f"{commit.author.name},,,,{commit.committer},,,,{commit.hash},,,,{commit.committer_date},,,,
                    {commit.project_name},,,,{commit.deletions},,,,{commit.insertions},,,,{commit.msg}")
data = pd.DataFrame(response, columns = ["Author"])
```

```python
data_modify = data["Author"].str.split(",,,,", expand = True)
```

```python
data_modify.rename(columns = {0: "author_name", 1: "commiter", 2: "id", 3: "commit_date", 4: "project", 5: "deletions", 6: "insertions", 7: "message"}, inplace = True)
```

```python
data_modify.to_csv("CSV/all_commits.csv")
```
